```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_credit = pd.read_csv('../data/raw/creditcard.csv')
```

```python
# --- PLOT 4: Class Imbalance ---
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df_credit)
plt.title("Class Distribution (Banking Data)")
plt.yscale('log') # Log scale to see the small bar
plt.savefig('credit_imbalance.png') # SAVE THIS IMAGE
plt.show()
```

```python
# --- PLOT 5: Amount vs Class ---
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Amount', data=df_credit)
plt.title("Transaction Amount by Class")
plt.ylim(0, 2000) # Zoom in to see the box
plt.savefig('credit_amount.png') # SAVE THIS IMAGE
plt.show()
```
